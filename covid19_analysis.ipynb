```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter sozlamasi
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('covid-data.csv')
df.head()
```

# 📍 Keraksiz ustunlarni olib tashlaymiz

Bu datasetda juda ko‘p ustunlar mavjud, lekin biz faqat eng asosiy ustunlar bilan ishlaymiz:

location – mamlakat nomi

date – sana

total_cases – jami holatlar

total_deaths – jami o‘limlar

population – aholi soni

```python
df = df[['location', 'date', 'total_cases', 'total_deaths', 'population']]
df.head()
```

# 📍 Sana ustunini datetime formatga o‘tkazamiz

```python
df['date'] = pd.to_datetime(df['date'])
```

# 📍 Ba’zi davlatlar bo‘yicha ma’lumotlarni ajratamiz
Masalan, O‘zbekiston, AQSH, Hindiston kabi davlatlarni tahlil **qilamiz**:

```python
countries = ['Uzbekistan', 'United States', 'India']
df_countries = df[df['location'].isin(countries)]
df_countries.tail()
```

# 📊 Har bir davlat bo‘yicha vaqt o‘tishi bilan jami holatlar o‘sishini ko‘rsatamiz

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_cases', hue='location')

plt.title("Davlatlar bo‘yicha jami COVID-19 holatlarining o‘sishi")
plt.xlabel("Sana")
plt.ylabel("Jami holatlar")
plt.legend(title='Davlat')
plt.tight_layout()
plt.savefig('images/covid_cases_over_time.png')
plt.show()
```

# 📊 Har bir davlat bo‘yicha jami o‘limlar o‘sishi

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_deaths', hue='location')

plt.title("Davlatlar bo‘yicha jami o‘limlar sonining o‘sishi")
plt.xlabel("Sana")
plt.ylabel("Jami o‘limlar")
plt.legend(title='Davlat')
plt.tight_layout()
plt.savefig('images/deaths.png')
plt.show()
```

# 📌 Aholiga nisbatan kasallanish foizi (total_cases / population)
Bu tahlil orqali har bir davlatda COVID-19 holatlari aholiga nisbatan qancha foizni tashkil qilganini ko‘ramiz.

✅ 1-Qadam: So‘nggi sanadagi ma'lumotlarni olish

```python
# Har bir davlat uchun eng so‘nggi mavjud ma'lumot
latest_df = df.sort_values('date').groupby('location').tail(1)
latest_df = latest_df[['location', 'total_cases', 'population']].dropna()
latest_df['cases_perc'] = latest_df['total_cases'] / latest_df['population']
```

```python
top10_perc = latest_df.sort_values(by='cases_perc', ascending=False).head(10)
top10_perc
```

```python
# Grafik
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_perc, x='cases_perc', y='location', palette='Blues_r')
plt.title("Top 10 davlat: Aholiga nisbatan eng yuqori COVID-19 holatlari (%)", fontsize=14)
plt.xlabel("Kasallanish foizi (0-1 orasida)")
plt.ylabel("Davlat")
plt.tight_layout()
plt.savefig("images/top10_perc.png")
plt.show()
```

# 📌 Eng yuqori o‘lim foiziga ega 10 ta davlat
Bu tahlilda har bir davlat uchun aholiga nisbatan o‘lim foizini hisoblaymiz (deaths_perc) va eng yuqori qiymatga ega 10 davlatni ko‘rsatamiz.

### 📦 1. So‘nggi total_deaths mavjud sanani topa

```python
# total_deaths mavjud va 0 dan katta bo‘lgan oxirgi sanani aniqlaymiz
latest_valid_date = df[df['total_deaths'].notna() & (df['total_deaths'] > 0)]['date'].max()
print("✅ So‘nggi mavjud to‘liq sana:", latest_valid_date)
```

### 🧹 2. Shu sanadagi toza ma'lumotlarni ajratamiz

```python
# So‘nggi sanadagi to‘liq malumotlarni olamiz
latest_df = df[df['date'] == latest_valid_date].copy()

# Faqat kerakli ustunlarni tanlaymiz
latest_df = latest_df[['location', 'total_cases', 'total_deaths', 'population']]

# NaN va aholisi 0 bo‘lganlar olib tashlanadi
latest_df = latest_df.dropna(subset=['total_deaths', 'population'])
latest_df = latest_df[latest_df['population'] > 0]
```

### 🧮 3. O‘lim foizini hisoblaymiz

```python
# Har bir davlatda o‘lim foizi (total_deaths / population)
latest_df['deaths_perc'] = latest_df['total_deaths'] / latest_df['population']
```

### 📊 4. Eng yuqori o‘lim foiziga ega 10 ta davlat

```python
# Eng yuqori o‘lim foiziga ega davlatlar
top10_deaths_perc = latest_df.sort_values(by='deaths_perc', ascending=False).head(10)

# Natijani ko‘rsatamiz
top10_deaths_perc[['location', 'total_deaths', 'population', 'deaths_perc']]
```

### 🎨 5. Grafik (vizual tahlil)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_deaths_perc, x='deaths_perc', y='location', palette='Reds_r')
plt.title("Eng yuqori o‘lim foiziga ega 10 ta davlat", fontsize=14)
plt.xlabel("O‘lim foizi (total_deaths / population)")
plt.ylabel("Davlat")
plt.tight_layout()
plt.savefig('images/top10_deaths_perc.png')
plt.show()
```
