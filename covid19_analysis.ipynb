```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Jupyter sozlamasi
sns.set(style="darkgrid")
```

```python
df = pd.read_csv('covid-data.csv')
df.head()
```

# ðŸ“ Keraksiz ustunlarni olib tashlaymiz

Bu datasetda juda koâ€˜p ustunlar mavjud, lekin biz faqat eng asosiy ustunlar bilan ishlaymiz:

location â€“ mamlakat nomi

date â€“ sana

total_cases â€“ jami holatlar

total_deaths â€“ jami oâ€˜limlar

population â€“ aholi soni

```python
df = df[['location', 'date', 'total_cases', 'total_deaths', 'population']]
df.head()
```

# ðŸ“ Sana ustunini datetime formatga oâ€˜tkazamiz

```python
df['date'] = pd.to_datetime(df['date'])
```

# ðŸ“ Baâ€™zi davlatlar boâ€˜yicha maâ€™lumotlarni ajratamiz
Masalan, Oâ€˜zbekiston, AQSH, Hindiston kabi davlatlarni tahlil **qilamiz**:

```python
countries = ['Uzbekistan', 'United States', 'India']
df_countries = df[df['location'].isin(countries)]
df_countries.tail()
```

# ðŸ“Š Har bir davlat boâ€˜yicha vaqt oâ€˜tishi bilan jami holatlar oâ€˜sishini koâ€˜rsatamiz

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_cases', hue='location')

plt.title("Davlatlar boâ€˜yicha jami COVID-19 holatlarining oâ€˜sishi")
plt.xlabel("Sana")
plt.ylabel("Jami holatlar")
plt.legend(title='Davlat')
plt.tight_layout()
plt.savefig('images/covid_cases_over_time.png')
plt.show()
```

# ðŸ“Š Har bir davlat boâ€˜yicha jami oâ€˜limlar oâ€˜sishi

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_countries, x='date', y='total_deaths', hue='location')

plt.title("Davlatlar boâ€˜yicha jami oâ€˜limlar sonining oâ€˜sishi")
plt.xlabel("Sana")
plt.ylabel("Jami oâ€˜limlar")
plt.legend(title='Davlat')
plt.tight_layout()
plt.savefig('images/deaths.png')
plt.show()
```

# ðŸ“Œ Aholiga nisbatan kasallanish foizi (total_cases / population)
Bu tahlil orqali har bir davlatda COVID-19 holatlari aholiga nisbatan qancha foizni tashkil qilganini koâ€˜ramiz.

âœ… 1-Qadam: Soâ€˜nggi sanadagi ma'lumotlarni olish

```python
# Har bir davlat uchun eng soâ€˜nggi mavjud ma'lumot
latest_df = df.sort_values('date').groupby('location').tail(1)
latest_df = latest_df[['location', 'total_cases', 'population']].dropna()
latest_df['cases_perc'] = latest_df['total_cases'] / latest_df['population']
```

```python
top10_perc = latest_df.sort_values(by='cases_perc', ascending=False).head(10)
top10_perc
```

```python
# Grafik
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_perc, x='cases_perc', y='location', palette='Blues_r')
plt.title("Top 10 davlat: Aholiga nisbatan eng yuqori COVID-19 holatlari (%)", fontsize=14)
plt.xlabel("Kasallanish foizi (0-1 orasida)")
plt.ylabel("Davlat")
plt.tight_layout()
plt.savefig("images/top10_perc.png")
plt.show()
```

# ðŸ“Œ Eng yuqori oâ€˜lim foiziga ega 10 ta davlat
Bu tahlilda har bir davlat uchun aholiga nisbatan oâ€˜lim foizini hisoblaymiz (deaths_perc) va eng yuqori qiymatga ega 10 davlatni koâ€˜rsatamiz.

### ðŸ“¦ 1. Soâ€˜nggi total_deaths mavjud sanani topa

```python
# total_deaths mavjud va 0 dan katta boâ€˜lgan oxirgi sanani aniqlaymiz
latest_valid_date = df[df['total_deaths'].notna() & (df['total_deaths'] > 0)]['date'].max()
print("âœ… Soâ€˜nggi mavjud toâ€˜liq sana:", latest_valid_date)
```

### ðŸ§¹ 2. Shu sanadagi toza ma'lumotlarni ajratamiz

```python
# Soâ€˜nggi sanadagi toâ€˜liq malumotlarni olamiz
latest_df = df[df['date'] == latest_valid_date].copy()

# Faqat kerakli ustunlarni tanlaymiz
latest_df = latest_df[['location', 'total_cases', 'total_deaths', 'population']]

# NaN va aholisi 0 boâ€˜lganlar olib tashlanadi
latest_df = latest_df.dropna(subset=['total_deaths', 'population'])
latest_df = latest_df[latest_df['population'] > 0]
```

### ðŸ§® 3. Oâ€˜lim foizini hisoblaymiz

```python
# Har bir davlatda oâ€˜lim foizi (total_deaths / population)
latest_df['deaths_perc'] = latest_df['total_deaths'] / latest_df['population']
```

### ðŸ“Š 4. Eng yuqori oâ€˜lim foiziga ega 10 ta davlat

```python
# Eng yuqori oâ€˜lim foiziga ega davlatlar
top10_deaths_perc = latest_df.sort_values(by='deaths_perc', ascending=False).head(10)

# Natijani koâ€˜rsatamiz
top10_deaths_perc[['location', 'total_deaths', 'population', 'deaths_perc']]
```

### ðŸŽ¨ 5. Grafik (vizual tahlil)

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top10_deaths_perc, x='deaths_perc', y='location', palette='Reds_r')
plt.title("Eng yuqori oâ€˜lim foiziga ega 10 ta davlat", fontsize=14)
plt.xlabel("Oâ€˜lim foizi (total_deaths / population)")
plt.ylabel("Davlat")
plt.tight_layout()
plt.savefig('images/top10_deaths_perc.png')
plt.show()
```
